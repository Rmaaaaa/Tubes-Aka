```python
import time
import matplotlib.pyplot as plt
from prettytable import PrettyTable

# Rekursif
def fibonacci_rekursif(n):
    if n <= 1:
        return n
    return fibonacci_rekursif(n - 1) + fibonacci_rekursif(n - 2)

# Iteratif
def fibonacci_iteratif(n):
    if n <= 1:
        return n
    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b

# Data Atlet
atlet = [
    {"nama": "Leo Rolly Carnando", "olahraga": "Bulutangkis", "cabang": "Ganda Putra", "kemenangan": 5},
    {"nama": "Fajar Alfian", "olahraga": "Bulutangkis", "cabang": "Ganda Putra", "kemenangan": 2},
    {"nama": "Farrel Armandio Tangkas", "olahraga": "Renang", "cabang": "Gaya Punggung", "kemenangan": 3},
    {"nama": "Glenn Victor Sutanto", "olahraga": "Renang", "cabang": "Gaya Kupu-Kupu", "kemenangan": 2},
    {"nama": "Jonatan Christie", "olahraga": "Bulutangkis", "cabang": "Tunggal Putra", "kemenangan": 4},
    {"nama": "Bobby Setiabudi", "olahraga": "Bulutangkis", "cabang": "Tunggal Putra", "kemenangan": 6},
    {"nama": "Michael Phelps", "olahraga": "Renang", "cabang": "Gaya Bebas", "kemenangan": 7},
    {"nama": "Caeleb Dressel", "olahraga": "Renang", "cabang": "Gaya Kupu-Kupu", "kemenangan": 5},
]

#untuk menampilkan tabel kemenangan berdasarkan olahraga dan mengurutkan kemenangan
def generate_tabel(cabang):
    tabel = PrettyTable()
    tabel.field_names = ["Atlet", "Olahraga", "Cabang", "Kemenangan"]
    filtered_atlet = [a for a in atlet if a["olahraga"].lower() == cabang.lower()]
    sorted_atlet = sorted(filtered_atlet, key=lambda x: x["kemenangan"], reverse=True)
    print_kemenangan_table(sorted_atlet)

    return sorted_atlet  # Mengembalikan data yang sudah diurutkan


#untuk menghitung waktu eksekusi rekursif dan iteratif berdasarkan kemenangan
def calculate_times(sorted_atlet):
    recursive_times = []
    iterative_times = []

    for a in sorted_atlet:
        kemenangan = a["kemenangan"]

        start = time.time()
        fibonacci_rekursif(kemenangan)
        recursive_times.append(time.time() - start)

        start = time.time()
        fibonacci_iteratif(kemenangan)
        iterative_times.append(time.time() - start)

    return recursive_times, iterative_times

#untuk mencetak tabel waktu eksekusi berdasarkan kemenangan
def print_execution_table(sorted_atlet, recursive_times, iterative_times):
    table = PrettyTable()
    table.field_names = ["Atlet", "Kemenangan", "Recursive Time (s)", "Iterative Time (s)"]
    for i in range(len(sorted_atlet)):
        table.add_row([sorted_atlet[i]["nama"], sorted_atlet[i]["kemenangan"],
                       f"{recursive_times[i]:.10e}", f"{iterative_times[i]:.10e}"])
    print("\nTabel Waktu Eksekusi (diurutkan berdasarkan Kemenangan tertinggi):")
    print(table)

def print_kemenangan_table(sorted_atlet):
    table = PrettyTable()
    table.field_names = ["Atlet", "Olahraga", "Cabang", "Kemenangan"]

    for atlet in sorted_atlet:
        table.add_row([atlet["nama"], atlet["olahraga"], atlet["cabang"], atlet["kemenangan"]])

    print()
    print(table)


# Visualisasi data
def plot_data(sorted_atlet, recursive_times, iterative_times):
    kemenangan_values = [a["kemenangan"] for a in sorted_atlet]

    plt.figure(figsize=(10, 5))
    plt.plot(kemenangan_values, recursive_times, marker='o', label="Waktu Rekursif", color='red')
    plt.plot(kemenangan_values, iterative_times, marker='x', label="Waktu Iteratif", color='blue')
    plt.title("Perbandingan Waktu Rekursif vs Iteratif Berdasarkan Kemenangan")
    plt.xlabel("Jumlah Kemenangan")
    plt.ylabel("Waktu (detik)")
    plt.grid(True)
    plt.legend()
    plt.show()

# Program utama
while True:
    try:
        cabang = input("Masukkan cabang olahraga (Bulutangkis/Renang): ").strip().lower()
        if cabang not in ["bulutangkis", "renang"]:
            print("Cabang tidak valid! Harap masukkan 'Bulutangkis' atau 'Renang'.")
            continue

        print(f"\nTabel Kemenangan Atlet {cabang.capitalize()}:")
        sorted_atlet = generate_tabel(cabang)

        recursive_times, iterative_times = calculate_times(sorted_atlet)

        print_execution_table(sorted_atlet, recursive_times, iterative_times)
        plot_data(sorted_atlet, recursive_times, iterative_times)

        break

    except ValueError:
        print("Masukkan nilai yang valid!")
```
